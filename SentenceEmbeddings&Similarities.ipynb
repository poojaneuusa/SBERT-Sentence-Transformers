```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ['the cat sits outside', 'the new moview is awesome', 'the new movie is really great', 'the dog bark on stangers','I just saw a new movie in a theatre and it was superbly picturised']
```

```python
embeddings = model.encode(sentences=sentences, convert_to_tensor=True)
```

```python
for sent,embed in zip(sentences, embeddings):
  print("Sentence:", sent)
  print("Len(Embeddings):", len(embed))
  # print("Embeddings: ", embed)
```

```python
cosine_scores = util.cos_sim(embeddings, embeddings)
```

```python
paraphrases = util.paraphrase_mining(model, sentences)
```

```python
for sim in paraphrases[0:10]:
  score, i, j = sim
  print(sentences[i], "<>" ,sentences[j], " --> ",score)
```
